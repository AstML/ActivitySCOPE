```python
import pandas as pd
import numpy as np
```

```python
orbits = pd.read_json("allcometels.json")
orbits.columns.tolist()
```

```python
orbits["Orbit_type"] + orbits["Provisional_packed_desig"]
```

```python
orbits = orbits[orbits["Orbit_type"].isin(["P","D","A"])]
orbits = orbits[orbits["Orbit_type"].isin(["A"])]
```

```python
orbits
```

```python
orbits['Tp'] = 2454466
```

```python
# calculate
orbits["a"] = orbits["Perihelion_dist"] / (1 - orbits["e"])

r = orbits['a']*(1+orbits['e']/2)
def add_vis(orb,param):
    orb['vis_'+str(param)] = (5*np.log10(np.maximum(r,1)*(np.maximum(r,1)-param))+orb['H']).astype(float)
    orb['vis_q_'+str(param)] = (5*np.log10(np.maximum(orb["Perihelion_dist"],1)*(np.maximum(orb["Perihelion_dist"],1)-param))+orb['H']).astype(float)

# These just happen to be ones that seem to work well, but it is open to tuning.
add_vis(orbits,0.6)
add_vis(orbits,-.1)
add_vis(orbits,-.2)
```

```python
mlcols = ['H', 'Peri', 'Node', 'i', 'Tp', 'vis_0.6', 'vis_q_-0.1', 'vis_q_-0.2', 'orbital_period_sync', 'Perihelion_direction_x_e', 'Perihelion_direction_y_e','Is_Past_Threshold']

orbits[mlcols]
```

```python
# read in all the lines from CmtObs.txt and output them to CmtObsFiltere.txt if they start with "    P"
with open("CmtObs.txt", "r") as infile, open("CmtObsFiltered.txt", "w") as outfile:
    for line in infile:
        if line.startswith("    A") or line.startswith("    P") or line.startswith("    D"):
            outfile.write(line)
```
